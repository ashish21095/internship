### Q1: Write a python program to scrape data for “Data Analyst” Job position in
“Bangalore” location. You have to scrape the job-title, job-location, company_name,
experience_required. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. first get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill,Designations,Companies” field and enter “Bangalore”
in “enter the location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note- All of the above steps have to be done in code. No step is to be done manually

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.naukri.com/")

input_f = driver.find_element_by_class_name("sugInp")
loc_f = driver.find_element_by_id('qsb-location-sugg')
input_f.send_keys('data analyst')
loc_f.send_keys('Bangalore')
search_bt = driver.find_element_by_class_name("btn")
search_bt.click()


### finding the element using xpath 
time.sleep(2)


main_container = driver.find_elements_by_xpath("//div[@class= 'info fleft']")

job_list = []
company_list = []
location_list = []
experience_list = []

for i in main_container[:10]:
    job_list.append(i.find_element_by_xpath(".//a").text)
    company_list.append(i.find_element_by_xpath(".//div[contains(@class,'companyInfo')]/a").text)
    location_list.append(i.find_element_by_xpath(".//ul/li[contains(@class,'location')]").text )
    experience_list.append(i.find_element_by_xpath(".//ul/li[contains(@class,'experience')]").text)
    
    
```

```python
import pandas as pd
pd.DataFrame(list(zip(job_list,location_list,company_list,experience_list)),columns=['Job-title','Job-location','Company_name','Experience_required'])
```

### Q2: Write a python program to scrape data for “Data Scientist” Job position in

manually

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.naukri.com/")

input_f = driver.find_element_by_class_name("sugInp")
loc_f = driver.find_element_by_id('qsb-location-sugg')
input_f.send_keys('Data Scientist')
loc_f.send_keys('Bangalore')
search_bt = driver.find_element_by_class_name("btn")
search_bt.click()


### finding the element using xpath 
time.sleep(2)


main_container = driver.find_elements_by_xpath("//div[@class= 'info fleft']")

job_list = []
company_list = []
location_list = []
experience_list = []
link_list = []
full_description = []
for i in main_container[:10]:
    job_list.append(i.find_element_by_xpath(".//a").text)
    company_list.append(i.find_element_by_xpath(".//div[contains(@class,'companyInfo')]/a").text)
    location_list.append(i.find_element_by_xpath(".//ul/li[contains(@class,'location')]").text )
    experience_list.append(i.find_element_by_xpath(".//ul/li[contains(@class,'experience')]").text)
    link_list.append(i.find_element_by_xpath(".//a").get_attribute('href'))
    
for i in link_list[:10]:
    driver.get(i)
    try:
        para =driver.find_element_by_class_name('dang-inner-html')
        full_description.append(para.text)
    except:
        para =driver.find_element_by_xpath("//div[@class='clearboth description']")
        full_description.append(para.text)
        

pd.DataFrame(list(zip(job_list,location_list,company_list,experience_list,full_description)),columns=['Job-title','Job-location','Company_name','Experience_required','Full_Description'])
```

### Q6 : Scrape data of first 100 sunglasses listings on flipkart.com. You have to
scrape four attributes:
1. Brand
2. Product Description
3. Price
4. Discount %
The attributes which you have to scrape is ticked marked in the below image.
To scrape the data you have to go through following steps:
1. Go to flipkart webpage by url https://www.flipkart.com/
2. Enter “sunglasses” in the search field where “search for products, brands and
more” is written and click the search icon
3. after that you will reach to a webpage having a lot of sunglasses. From this page
you can scrap the required data as usual.
4. after scraping data from the first page, go to the “Next” Button at the bottom of
the page , then click on it
5. Now scrape data from this page as usual
6. repeat this until you get data for 100 sunglasses.
Note that all of the above steps have to be done by coding only and not manually.


```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.flipkart.com/')

time.sleep(2)
pro_count = 0
driver.find_element_by_xpath('//button[text()="✕"]').click()
driver.find_element_by_xpath('//input[@name="q"]').send_keys('sunglasses')
driver.find_element_by_xpath('//button[@type="submit"]').click()

time.sleep(2)


brand_list=[]
product_desc=[]
price = []
discount = []

while pro_count < 100:
    main_container= driver.find_elements_by_xpath('//div[contains(@data-id,"SGL")]')
    for i in main_container:
        try:
            brand_list.append(i.find_element_by_xpath('.//div[@class="_2WkVRV"]').text)
            
        except:
            brand_list.append('')
        try:
            product_desc.append(i.find_element_by_xpath('.//div[@class="_2B099V"]/a[@class="IRpwTa"]').text)
        except:
            product_desc.append('')
        try:      
            price.append(i.find_elements_by_xpath('.//div[@class="_25b18c"]/div')[0].text)
        except:
            price.append('')
        try:
            discount.append(i.find_elements_by_xpath('.//div[@class="_25b18c"]/div')[2].text)
        except:
            discount.append('')
        pro_count += 1    
        if pro_count == 100:
            break;
            
    driver.find_element_by_xpath('//nav/a[@class="_1LKTO3" ]/span[text() = "Next"]').click()
    time.sleep(1)
```

```python
pd.DataFrame(list(zip(brand_list,product_desc,price,discount)),columns=['Brand','Product_desc','Price','Discount'])
```
